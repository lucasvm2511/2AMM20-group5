```python
import joblib
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_qualities1 = np.load("artifacts/all_qualities.npy")
all_qualities2 = np.load("artifacts/all_qualities_2.npy")
all_qualities = np.vstack([all_qualities1, all_qualities2]).mean(0)
all_qualities_rgb1 = np.load("artifacts/all_qualities_rgb.npy")
all_qualities_rgb2 = np.load("artifacts/all_qualities_rgb_2.npy")
all_qualities_rgb = np.vstack([all_qualities_rgb1, all_qualities_rgb2]).mean(0)
```

```python
all_sizes1 = np.load("artifacts/all_sizes.npy")
all_sizes2 = np.load("artifacts/all_sizes_2.npy")
all_sizes = np.vstack([all_sizes1, all_sizes2]).mean(0) / 13
all_sizes_rgb1 = np.load("artifacts/all_sizes_rgb.npy")
all_sizes_rgb2 = np.load("artifacts/all_sizes_rgb_2.npy")
all_sizes_rgb = np.vstack([all_sizes_rgb1, all_sizes_rgb2]).mean(0) / 3
```

```python
sizes_qualities = list(zip(all_sizes, all_qualities))
sorted_sizes_qualities = sorted(sizes_qualities, key=lambda x: x[0])
sorted_sizes_qualities = list(filter(lambda x: x[0] <= 40, sorted_sizes_qualities))
sorted_sizes, sorted_qualities = zip(*sorted_sizes_qualities)
sorted_sizes = np.array(list(sorted_sizes))
sorted_qualities = np.array(list(sorted_qualities))

sizes_qualities_rgb = list(zip(all_sizes_rgb, all_qualities_rgb))
sorted_sizes_qualities_rgb = sorted(sizes_qualities_rgb, key=lambda x: x[0])
sorted_sizes_qualities_rgb = list(filter(lambda x: x[0] <= 40, sorted_sizes_qualities_rgb))
sorted_sizes_rgb, sorted_qualities_rgb = zip(*sorted_sizes_qualities_rgb)
sorted_sizes_rgb = np.array(list(sorted_sizes_rgb))
sorted_qualities_rgb = np.array(list(sorted_qualities_rgb))
```

```python
plt.plot(sorted_sizes, sorted_qualities)
plt.plot(sorted_sizes_rgb, sorted_qualities_rgb)
plt.axvline(x=16, color="r", linestyle="--")
plt.xlabel("BPP per channel")
plt.ylabel("PSNR value")
plt.grid(True)
plt.savefig("artifacts/coin_plot.png")
plt.show()
```
